```r
data_all <- train %>%
    group_by(iyear) %>% 
   summarise(count = n())

head(data_all)
```

```r
# changes in the trend of #attacks Vs Year
ggplot(data_all, aes(x=iyear, y=count)) + 
  geom_line() +
  labs(title="Number of Attacks per Year for major", x="Year", y="Number of Attacks")
```

```r
model_lm <- lm(count ~ iyear , data = data_all)
```

This output is from a linear regression model where count is the dependent variable and iyear is the independent variable. Here are some interpretations of the different parts of the output:

Call: This shows the call to the lm() function that was used to create the model.
Residuals: This gives information about the residuals (or errors) of the model, including their minimum, maximum, median, and quartiles.
Coefficients: This shows the estimates for the intercept and slope coefficients of the model. The estimate for the intercept is -355058.46, which means that the expected count when iyear is zero (which is not a meaningful value in this case) is -355058.46. The estimate for the slope is 180.05, which means that for every one unit increase in iyear, the expected count increases by 180.05. Both the intercept and slope estimates are statistically significant, as indicated by the *** in the Pr(>|t|) column.
Residual standard error: This is the estimate of the standard deviation of the error term in the model.
Multiple R-squared: This is the proportion of variance in count that is explained by iyear. In this case, the value is 0.4409, which means that iyear explains about 44.09% of the variance in count.
Adjusted R-squared: This is a modified version of the R-squared that takes into account the number of variables in the model and the sample size. In this case, the adjusted R-squared is 0.4285.
F-statistic: This is the ratio of the explained variance to the unexplained variance in the model, and is used to test the overall significance of the model. In this case, the F-statistic is 35.49, which is statistically significant with a very low p-value.
p-value: This is the probability of observing a test statistic as extreme as the one calculated, assuming that the null hypothesis (no relationship between count and iyear) is true. In this case, the p-value is very small (3.61e-07), which means that we can reject the null hypothesis and conclude that there is a statistically significant relationship between count and iyear.
